```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 1. Basic Matplotlib visualization:

#### 1.1. Bar plot:

```python
x = np.array(['Q1', 'Q2', 'Q3','Q4'])
y = np.array([ 234.0, 254.7, 144.6, 317.6])
plt.bar(x,y,color = 'purple')
plt.title('BAR CHART')
plt.show()
```

#### 1.2. Histogram:

```python
x = np.random.randn(10000)
#randn generates a normal distribution
plt.hist(x,bins=20,color='green',density=True)
#bins are the pockets in the histogram, it's important to choose the right number
#density = True does frequency/total_number_of_observations. it's a value [0,1]
#density = False puts frequency in y axis
plt.title('HISTOGRAM')
plt.show()
```

```python
x = np.random.randn(10000)
plt.hist(x,bins=100,color='blue',density=False,alpha=0.3)
#alpha is related to the transparency of the histogram 
plt.title('HISTOGRAM')
plt.show()
```

#### 1.3. Box plot:

```python
x = np.random.randn(10000)*10+3
y = np.random.randn(10000)*10+5
z = np.random.randn(10000)*10+1
plt.boxplot([x,y,z],0, vert=False)
#vert = Falses puts the plot in the horizontal
plt.title('BOX PLOTS')
plt.show()
```

```python
plt.boxplot([x,y,z],0,vert=False)
plt.title('BOX PLOTS')
plt.show()
```

#### 1.4. Line plot:

```python
x = np.linspace(0,10,100)
#np.linspace creates a list between 0 and 10 with 100 elements in total at the same distance to each other
y = np.sin(x)
plt.plot(x,y,color='orange',linewidth=7, linestyle='-.')  # linestyle = '-'
#linewidth relalate to thickness of line        
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='blue',linestyle=':',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='red',linestyle='-.',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='green',linestyle='--',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='purple',linestyle='steps',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='purple',linestyle='-',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.xlim([0,5])                                     # Horizontal axis limits.
plt.ylim([-2,+2])                                   # Vertical axis limits.
plt.show()
```

#### 1.5. Scatter plot 1:

Scatter plots with plot() function and linestyle ='none':

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='red',marker='o',linestyle='none',markersize=20, alpha=0.5)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='#FF8C00',marker='o',linestyle='none',markersize=5, alpha=0.5)# RGB color.
plt.plot(x,y,color='#FF8C00',marker='o',markersize=5, alpha=0.5)     #without linestyle=none there i a light connecting the dots
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,marker='o',linestyle='none',markersize=15,markerfacecolor='yellow',markeredgecolor='purple',markeredgewidth=2) #fills the O with yellow inside, purple outside with linewidth = 2
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```
